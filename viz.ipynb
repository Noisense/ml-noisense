```python
import pandas as pd
import pymysql
import matplotlib.pyplot as plt

host = 'localhost'
user = 'root'
password = 'noisense'
database = 'noisense'
sql_query = 'SELECT * FROM extract_files'

conn = pymysql.connect(host=host, user=user, password=password, database=database)

df = pd.read_sql(sql_query, conn)

conn.close()
```

```python
plt.figure(figsize=(8, 6))
value = df['label'].value_counts()
ax = value.plot(kind='bar', color='skyblue')
for i, v in enumerate(value):
    ax.text(i, v + 0.1, str(v), ha='center', va='bottom')
plt.title('Jumlah Data untuk Setiap Label')
plt.xlabel('Label')
plt.ylabel('Jumlah')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
value = df['date_recorded'].value_counts()
ax = value.plot(kind='bar', color='skyblue')
for i, v in enumerate(value):
    ax.text(i, v + 0.1, str(v), ha='center', va='bottom')
plt.title('Jumlah File untuk Setiap Label')
plt.xlabel('Tanggal')
plt.ylabel('Jumlah')
plt.show()
```
